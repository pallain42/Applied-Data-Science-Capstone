```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
import folium, math
from folium.plugins import MousePosition

# --- choose the launch site you want to measure from ---
# If your notebook already defines these, keep them; otherwise set by site name:
site_name = "CCAFS LC-40"  # <- change if you want a different site
launch_row = spacex_df[spacex_df['Launch Site'] == site_name].iloc[0]
launch_site_lat = float(launch_row['Lat'])
launch_site_lon = float(launch_row['Long'])

# --- add a live coordinate readout so you can pick a coastline point ---
MousePosition(position='topright',
              separator=' | ',
              prefix='Lat/Lon:').add_to(launch_map)

# After reading coords from the map HUD, paste them here:
coast_lat, coast_lon = 28.56367, -80.57163   # <-- REPLACE with your picked coastline coords

# --- distance function (km) ---
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    phi1, phi2 = math.radians(lat1), math.radians(lat2)
    dphi = math.radians(lat2 - lat1)
    dlmb = math.radians(lon2 - lon1)
    a = math.sin(dphi/2)**2 + math.cos(phi1)*math.cos(phi2)*math.sin(dlmb/2)**2
    return 2*R*math.asin(math.sqrt(a))

distance_km = haversine(launch_site_lat, launch_site_lon, coast_lat, coast_lon)

# --- add labeled marker at the coastline point ---
folium.Marker(
    [coast_lat, coast_lon],
    icon=folium.DivIcon(
        icon_size=(200,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size:12px;color:#d35400;"><b>{site_name}</b><br>≈ {distance_km:.2f} km to coast</div>'
    ),
    tooltip="Coast point"
).add_to(launch_map)

# --- draw line from launch site to coast point ---
folium.PolyLine(
    locations=[(launch_site_lat, launch_site_lon), (coast_lat, coast_lon)],
    weight=2,
    color='blue'
).add_to(launch_map)

# --- optional: add a small marker at the launch site for clarity ---
folium.CircleMarker(
    location=[launch_site_lat, launch_site_lon],
    radius=5, color='black', fill=True, fill_opacity=1.0,
    tooltip=f"{site_name} (launch site)"
).add_to(launch_map)

# --- fit view to both points and display ---
launch_map.fit_bounds([(launch_site_lat, launch_site_lon), (coast_lat, coast_lon)])
launch_map
```

```python
import folium, math
from folium.plugins import MousePosition
from folium import DivIcon

# ---- 1) Pick the launch site (change the name if you want a different site) ----
site_name = "CCAFS LC-40"
if 'Class' in spacex_df.columns and 'class' not in spacex_df.columns:
    spacex_df.rename(columns={'Class':'class'}, inplace=True)

site = spacex_df[spacex_df['Launch Site'] == site_name].iloc[0]
launch_site_lat = float(site['Lat'])
launch_site_lon = float(site['Long'])

# ---- 2) Create/attach to the map ----
center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
launch_map = folium.Map(location=center, zoom_start=9, tiles='cartodbpositron')
MousePosition(position='topright', separator=' | ', prefix='Lat/Lon:').add_to(launch_map)

# Mark the launch site
folium.CircleMarker([launch_site_lat, launch_site_lon], radius=6, color='black', fill=True,
                    fill_opacity=1.0, tooltip=f"{site_name} (launch site)").add_to(launch_map)

# ---- 3) Distance helper ----
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    import math as _m
    p1, p2 = _m.radians(lat1), _m.radians(lat2)
    dphi = _m.radians(lat2 - lat1)
    dlmb = _m.radians(lon2 - lon1)
    a = _m.sin(dphi/2)**2 + _m.cos(p1)*_m.cos(p2)*_m.sin(dlmb/2)**2
    return 2*R*_m.asin(_m.sqrt(a))

# ---- 4) FILL THESE with coords you read from the map HUD ----
coast_lat,  coast_lon  = 28.56367, -80.57163   # coastline point (example)
rail_lat,   rail_lon   = 28.57210, -80.64800   # nearest railway (replace)
hwy_lat,    hwy_lon    = 28.55750, -80.80750   # nearest highway (replace)
city_lat,   city_lon   = 28.61222, -80.80755   # nearest city center (replace)

features = {
    "Coastline": ((coast_lat, coast_lon),  "#d35400", "blue"),
    "Railway":   ((rail_lat, rail_lon),   "#8e44ad", "purple"),
    "Highway":   ((hwy_lat,  hwy_lon),    "#16a085", "green"),
    "City":      ((city_lat, city_lon),   "#2980b9", "navy"),
}

# ---- 5) Plot all proximities with labels + lines ----
bounds = [(launch_site_lat, launch_site_lon)]
for name, ((lat, lon), label_color, line_color) in features.items():
    dist_km = haversine(launch_site_lat, launch_site_lon, lat, lon)

    folium.Marker(
        [lat, lon],
        icon=DivIcon(html=f'<div style="font-size:12px;color:{label_color};"><b>{name}</b><br>≈ {dist_km:.2f} km</div>')
    ).add_to(launch_map)

    folium.PolyLine([(launch_site_lat, launch_site_lon), (lat, lon)],
                    color=line_color, weight=2).add_to(launch_map)

    bounds.append((lat, lon))

# Fit view to show everything
launch_map.fit_bounds(bounds)
launch_map
```

```python
**Findings:**

- **Railways:** The launch site is within ~0.6 km of a railway, meaning it is well connected for transporting heavy equipment.  
- **Highways:** The site is also close to a highway (~22 km), providing road access for staff and cargo.  
- **Coastline:** The site is extremely close to the coastline (~0.6 km), which allows safe rocket trajectories over the ocean.  
- **Cities:** The nearest city (Titusville) is about 23 km away, showing that launch sites are kept at a safe distance from population centers.  

Overall, launch sites are selected to maximize safety (close to coast, away from cities) while maintaining good access to logistics (railway and highway).
```
